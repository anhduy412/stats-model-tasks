## Data preprocessing

### Get the neccessary data

```python
retail_sales_data = retail_sales_data[['product_key', 'order_quantity', 'unit_price', 'unit_cost', 'unit_discount', 'sales_amount']]
retail_sales_data
```

```python
partner_sales_data = partner_sales_data[['product_key', 'order_quantity', 'unit_price', 'unit_cost', 'unit_discount', 'sales_amount']]
partner_sales_data
```

```python
ecommerce_sales_data = ecommerce_sales_data[['product_key', 'order_quantity', 'unit_price', 'unit_cost', 'unit_discount', 'sales_amount']]
ecommerce_sales_data
```

```python
online_sales_data = online_sales_data[['product_key', 'order_quantity', 'unit_price', 'unit_cost', 'unit_discount', 'sales_amount']]
online_sales_data
```

```python
# Concatenate all DataFrames into a single DataFrame
sales_data = pd.concat([retail_sales_data, partner_sales_data, ecommerce_sales_data, online_sales_data])

# Group by "product_key" and sum the other columns
sales_data = sales_data.groupby('product_key', as_index=False).sum()
sales_data
```

```python
sales_data['profit'] = (sales_data['unit_price'] - sales_data['unit_discount'] - sales_data['unit_cost']) * sales_data['order_quantity']
sales_data['profit_margin'] = (sales_data['profit'] / sales_data['sales_amount']) * 100
sales_data
```

# Task 3: Visualise the data

# Task 4. Create the model

```python
top_10_products = sales_data.nlargest(10, 'profit')
top_10_products
```

```python
# X = sales_data[['sales_amount', 'profit_margin']]
# X = sales_data[['unit_price', 'unit_cost', 'unit_discount']]
X = top_10_products[['unit_price', 'unit_cost', 'unit_discount', 'sales_amount', 'profit_margin']]
y = top_10_products['profit']
```

```python
# Scale the features for better model performance
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Build the linear regression model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_scaled, y)
```

```python
# Predict the optimal allocation for all products
# X_all = sales_data[['sales_amount', 'profit_margin']] 
# X_all = sales_data[['unit_price', 'unit_cost', 'unit_discount']]
X_all = sales_data[['unit_price', 'unit_cost', 'unit_discount', 'sales_amount', 'profit_margin']]
X_all_scaled = scaler.transform(X_all)
sales_data['predicted_profit'] = model.predict(X_all_scaled)
```

```python
# Calculate the actual allocation for each product based on the 100 billion fund
total_fund = 100000000000
sales_data['allocation_amount'] = (sales_data['predicted_profit'] / sales_data['predicted_profit'].sum()) * total_fund
```

```python
# Display the top 10 products and the percentage of money to be invested
top_10_products = sales_data.nlargest(10, 'allocation_amount')
top_10_products['allocation_percentage'] = (top_10_products['allocation_amount'] / total_fund) * 100
top_10_products[['product_key', 'allocation_amount', 'allocation_percentage']]
```
